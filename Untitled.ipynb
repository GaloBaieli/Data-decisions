```python
df=pd.read_csv('Data/datasets/cwurData.csv')
df.head(1)
```

```python
def create_category (ranking):
    if (ranking>=1)&(ranking<=100):
        return 'First tier'
    elif(ranking>100)&(ranking<=200):
        return 'Second tier'
    elif(ranking>200)&(ranking<=300):
        return 'Third tier'
    return 'Other top university'
```

```python
df['Rank_level']=df['world_rank'].apply(lambda y:create_category(y))
df.head(2)
```

```python
myPivot=df.pivot_table(values='score',
                       index='country',
                       columns='Rank_level'
                      )
myPivot.head()
```
